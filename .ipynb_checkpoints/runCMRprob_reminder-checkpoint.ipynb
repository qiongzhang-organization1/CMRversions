### Effectiveness of reminders by serial position

```python
num = len(CMR_sp)
maxs = []
mins = []
rnds = []
accs_pos = [[] for i in range(ll)]
for i in range(num):
    acc = [np.mean(x) for x in accs[i]]
    maxs.append(np.max(acc))
    mins.append(np.min(acc))
    ind = random.randint(0,len(acc)-1)
    rnds.append(acc[ind])
    for j,x in enumerate(reminders[i]):
        accs_pos[x].append(acc[j])
    
```

```python
plt.rcParams['figure.figsize'] = (10,5.5)
mus = [np.mean(x) for x in accs_pos]
sems = [np.std(x)/np.sqrt(len(x)-1) for x in accs_pos]
plt.bar(range(ll), mus, width=1, yerr=sems,color='grey')
#plt.xticks([1,3,5], ('Worst reminder','Random reminder','Best reminder'))
plt.ylabel('Additional items recalled')
plt.show()
print(np.mean(acc))
```

```python
 plt.rcParams['figure.figsize'] = (5.5,5.5)
y1c = np.mean(mins)
error1c = np.std(mins)/np.sqrt(len(mins)-1)
y1d = np.mean(rnds)
error1d = np.std(rnds)/np.sqrt(len(rnds)-1)
y1g = np.mean(maxs)
error1g = np.std(maxs)/np.sqrt(len(maxs)-1)
# non-parametric
stat, p = mannwhitneyu(mins, maxs)
print('Statistics=%.3f, p=%.3f' % (stat, p))
plt.bar([1,3,5], [y1c,y1d,y1g], width=1, yerr=[error1c,error1d,error1g],color='grey')
plt.xticks([1,3,5], ('Worst reminder','Random reminder','Best reminder'))
plt.ylabel('Additional items recalled')
plt.show()
print(np.mean(acc))
```
